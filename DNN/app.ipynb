```python
import pandas as pd
```

```python
# Load the dataset
file_path = '../messages.csv'
data = pd.read_csv(file_path)
data
```

```python
data.head()
```

```python
# Drop the columns that are not useful
data['subject'].fillna('', inplace=True)

# Combine the subject and message columns
data['text'] = data['subject'] + ' ' + data['message']

# Print the first few rows of the combined text
print(data[['text', 'label']].head())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer

# Extract the text and label columns
X_train, X_test, y_train, y_test = train_test_split(data['text'], data['label'], random_state=0)

# Create TF-IDF vectorziers

ftidf_vectorizer = TfidfVectorizer(max_features=1000)
X_train_tfidf = ftidf_vectorizer.fit_transform(X_train)
X_test_tfidf = ftidf_vectorizer.transform(X_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# Create a sequential model with three layers of Dense neurons
model = Sequential([
    Dense(512, input_dim=1000, activation='relu'),
    Dense(256, activation='relu'),
    Dense(1, activation='sigmoid')
])

model
```

```python
# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy',metrics=['accuracy'])

# Check the shape of the input features and labels
print("Shape of X_train:", X_train_tfidf.shape)
print("Shape of y_train:", y_train.shape)

# Check model's input shape
print("Model's expected input shape:", model.input_shape)


X_train_tfidf = X_train_tfidf.toarray()  # Convert training data
X_test_tfidf = X_test_tfidf.toarray()    # Convert test data


# Check the shape of the input features and labels
print("Shape of X_test:", X_test_tfidf.shape)



```

```python
# Train the model
history = model.fit(X_train_tfidf, y_train, epochs=310, batch_size=8,validation_data=(X_test_tfidf, y_test))
```
